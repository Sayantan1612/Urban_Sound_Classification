## **OBTAINING TRAINING DATA AND CONVERT TO ONE-HOT-ENCODE**

```
train = pd.read_csv(precedeString_Train + '/train.csv') #READING TRAINING CSV FILE
```

# Mel Frequency Cepstral Coefficients (Speech Processing)


## 1.    First the audio signal is divided into Time Frames.
## 2.    Cepstrum is the information of rate of change in spectral bands.
## 3.    Filter Bank is basically the frequencey scaling by which we can absorb both low and high frequencies
##4.    Fourier transform are the Feature vectors
    



```
def parser_Train(row):   
    file_name = precedeString_Train + '/'+str(row.ID) + '.wav' #MAKING OUT THE FILE NAME FOR EXTRACTION
    try:
      X, sample_rate = librosa.load(file_name,res_type='kaiser_fast')  #LOAD AND EXTRACT FEATURES
      mfccs = np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40).T,axis=0) #MFCCS FEATURE USED TO ANALYZE THE FEATURES
    except Exception as e:
      print("Error encountered while parsing file: ", file_name)
      return None, None
    feature = mfccs
    
    label = row.CLASSES
    print(file_name)
    return pd.Series([feature,label])
```

```
temp = train.apply(parser_Train, axis=1) #PARSING THROUGH ALL FILE NAMES
temp.columns = ['feature','label']
```

```
X = np.array(temp.feature.tolist())
y = np.array(temp.label.tolist())
print(y[0])
print(y[500])
print(y[1000])
print(y[799])

lb2 = LabelEncoder()
y=lb2.fit_transform(y)
print(y[0])
print(y[500])
print(y[1000])
print(y[799])

lb = LabelEncoder()  #LABELENCODER IS USED TO CONVERT STRINGS TO UNIQUE NUMBERS
y = np_utils.to_categorical(y) #CONVERTING IN THE FORM OF ONE-HOT -ENCOCDE
```

## **OBTAINING TESTING DATA AND CONVERT TO ONE-HOT-ENCODE**

```
validate = pd.read_csv(precedeString_Val + '/test.csv') #READING TEST CSV FILE 
```

```
def parser_Val(row):    #SAME AS DONE IN CASE OF TRAINING, LOADING OF FILE AND EXTRACTING FEATURES USING MFCC
    file_name = precedeString_Val + '/' + row.ID + '.wav'
    try:
      # here kaiser_fast is a technique used for faster extraction
      val_X, sample_rate = librosa.load(file_name,res_type='kaiser_fast') 
      # we extract mfcc feature from data
      mfccs = np.mean(librosa.feature.mfcc(y=val_X, sr=sample_rate, n_mfcc=40).T,axis=0) 
    except Exception as e:
      print("Error encountered while parsing file: ", file_name)
      return None, None
    feature = mfccs
    label = row.CLASSES
    print(file_name)
    return pd.Series([feature,label])
```

```
temp_Val = validate.apply(parser_Val, axis=1)
temp_Val.columns = ['feature', 'label']
```

```
val_X = np.array(temp_Val.feature.tolist())
val_Y = np.array(temp_Val.label.tolist())


lb2 = LabelEncoder()
val_Y=lb2.fit_transform(val_Y)

val_Y = np_utils.to_categorical(val_Y) #CONVERT TEST_Y TO ONE HOT ENCODE
print(val_Y.shape)
```

## **CREATE SEQUENTIAL MODEL USING KERAS**

```
num_labels = y.shape[1]
filter_size = 2    #FOR DECIDING NO. OF NEURONS IN OUTPUT LAYER
```

![alt text](http://d33wubrfki0l68.cloudfront.net/0b32029a0c0f19a062d57dd101d3c85148e505e2/d4843/images/tasks/challenge2017/task2_overview.png)

```
#input_layer=Input(shape=(X.shape[1:]))
model = Sequential()

model.add(Dense(50, input_shape = (40,)))
model.add(Activation('relu'))

model.add(Dense(50))
model.add(Activation('relu'))


model.add(Dense(50))
model.add(Activation('relu'))


model.add(Dense(50))
model.add(Activation('relu'))

model.add(Dense(num_labels))
model.add(Activation('sigmoid'))



model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X,y, epochs=200,validation_split=0.1)
```

```
model.summary()
```

## **PREDICTING BY PASSING THE val_X THROUGH THE MODEL AND THEN CREATING CONFUSION MATRIX**

```
loss, accuracy = model.evaluate(val_X, val_Y,verbose=0)  # Evaluate the model
print('Accuracy :%0.3f'%accuracy)
print('Loss:%0.3f'%loss)
```

```
pred_labels=model.predict(val_X)
cm=confusion_matrix(val_Y.argmax(axis=1),pred_labels.argmax(axis=1))
print(cm)

#babycry
#glassbreak
#gunshot
#others
```

```
f1_score(val_Y,pred_labels.round(),average=None)
```
